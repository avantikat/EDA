```python
df.country = pd.read_excel('/content/Country-Code.xlsx')
df.country.head()
```

```python
final_df=pd.merge(df,df.country , on='Country Code', how = 'left' )
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
county_val=final_df.Country.value_counts().values
```

```python
plt.pie(county_val[:3],labels= country_names[:3], autopct='%1.2f%%')  #top 3 country
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
ratings
```

```python
sns.barplot(x ="Aggregate rating",y="Rating_count" ,data =ratings)
```

```python
import matplotlib 
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x ="Aggregate rating",y="Rating_count" ,data =ratings)
```

```python
sns.barplot(x ="Aggregate rating",y="Rating_count" ,hue='Rating color' ,data =ratings,palette=['White','RED','Orange','Yellow','Green','green'])
```

```python
sns.countplot(x='Rating color',data =ratings,palette=['White','RED','Orange','Yellow','Green','green'])
```
